```python
# Cria Pastas iniciais e move arquivos uplodados para a pasta de Input

import os
from os import listdir

os.system("mkdir Input")
os.system("mkdir Rotacionado")
os.system("mkdir Ocr")
os.system("mkdir Output")

input_dir = "/content/"
output_dir = "/content/Input/"

for x in listdir(input_dir):
  if not x.endswith(".pdf"):
    continue
  os.system("mv '{}' {}" .format(input_dir + x, output_dir))
```

```python
#Girar todos os PDFs de uma pasta e salvar em outra

import os
from PyPDF2 import PdfFileReader, PdfFileWriter
from os import listdir

input_dir = "/content/Input/"
output_dir = "/content/Rotacionado/"

rotation = 90 #Define a rotação sendo 0° a posisão de um pdf em pé e os valores saumentam no sentido horário

# count = 0   #se quiser ir adicionando números no nome do pdf

for x in listdir(input_dir):
    if not x.endswith('.pdf'):
        continue
    pdf_in = open(input_dir + x, 'rb')
    pdf_reader = PdfFileReader(pdf_in)
    pdf_writer = PdfFileWriter()
    for pagenum in range(pdf_reader.numPages):
        page = pdf_reader.getPage(pagenum)
        page.rotateClockwise(rotation)
        pdf_writer.addPage(page)
    pdf_out = open(output_dir + x, 'wb')                  #Mantém o nome original
    # pdf_out = open(output_dir + x + str(count), 'wb')   #é possivel concatenar ou alterar os nomes 
    pdf_writer.write(pdf_out)
    pdf_out.close()
    pdf_in.close()
    # count += 1                                           #Incrementa o contador
```

```python
#Faz o OCR de todos os arquivos numa pasta e salva em outra

import ocrmypdf
from os import listdir

input_dir = "/content/Input/"   
output_dir = "/content/Ocr/"

for x in listdir(input_dir):
    if not x.endswith('.pdf'):
        continue
    ocrmypdf.ocr(input_dir + x, output_dir + x)
```

```python
#Juntar todos os PDFs de uma pasta e salvar em outra

import os
from PyPDF2 import PdfFileReader, PdfFileWriter, PdfFileMerger
from os import listdir

input_dir = "/content/Input/"
output_dir = "/content/Output/"

pdf_merger = PdfFileMerger()

for x in listdir(input_dir):
    if not x.endswith('.pdf'):
        continue
    pdf_in = open(input_dir + x, 'rb')
    pdf_reader = PdfFileReader(pdf_in)
    pdf_merger.append(pdf_reader)

pdf_merger.write("document-output.pdf")
```
